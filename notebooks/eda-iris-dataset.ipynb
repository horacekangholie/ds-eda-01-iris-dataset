### EDA - Iris Dataset

##### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

##### Load Dataset

```python
# Load the Iris dataset and create a DataFrame
iris = load_iris(as_frame=True)
X_df = iris.data
y_df = iris.target
df_data = pd.concat([X_df, y_df], axis=1)     # df_data = iris.frame    # Alternative
```

##### Histogram

```python
# Prepare the list of feature names from the dataset
feature_names = iris.feature_names

# Create subplots: 1 row, 4 columns (one for each feature)
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(15, 4))

# Loop through each feature and plot histogram with KDE
for i, feature in enumerate(feature_names):
    axes[i].set_title(feature)
    sns.histplot(data=df_data, x=feature, ax=axes[i], kde=True)

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

##### Countplot

```python
ax = sns.countplot(data=df_data, x='target', hue='target', palette='tab10')
plt.legend(title="target", labels=iris.target_names, bbox_to_anchor=(1.05, 1), loc='upper left')
for container in ax.containers:
    ax.bar_label(container)
```

##### Pairplot

```python
sns.pairplot(df_data, hue='target', diag_kind='kde', palette='tab10')
```

##### Heatmap

```python
# Calculate the correlation matrix for all numeric columns
corr = df_data.corr()

# hide duplicate values of upper triangle matrix
mask = np.triu(np.ones_like(corr, dtype=np.bool_))

# Draw the map
sns.heatmap(corr, square=True, annot=True, mask=mask, cmap="RdBu_r")
```

##### Scatter Plot

```python
sns.lmplot(x="petal length (cm)", y="petal width (cm)", hue='target', data=df_data, fit_reg=False, legend=False)
plt.legend(title='Target', loc='upper left', labels=iris.target_names)
```
