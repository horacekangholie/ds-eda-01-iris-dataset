### EDA - Iris Dataset

##### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

##### Load Dataset

```python
# Load dataset
iris = load_iris(as_frame=True)

X_df = iris.data
y_df = iris.target    
```

```python
# df_data = iris.frame    # Alternative
df_data = pd.concat([X_df, y_df], axis = 1)
df_data
```

#### Histogram

```python
fig, axes = plt.subplots(nrows=1, ncols=4)
fig.set_size_inches(15, 4)
sns.histplot(x="sepal length (cm)", data=df_data, ax=axes[0], kde=True)
sns.histplot(x="sepal width (cm)", data=df_data, ax=axes[1], kde=True)
sns.histplot(x="petal length (cm)", data=df_data, ax=axes[2], kde=True)
sns.histplot(x="petal width (cm)", data=df_data, ax=axes[3], kde=True)
```
